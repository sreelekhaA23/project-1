```python
x=data.drop("price",axis=1)
```

```python
x
```

```python
y=data['price']
```

```python
y
```

```python
pip install sklearn
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()   #creating object of the linear regression class
model.fit(x,y)   # model training
```

```python
model.coef_    #finding values of c
```

```python
model.intercept_
```

```python
y_pred=model.predict(x)    #storing the predicted y value in y_pred
y_pred
```

```python
pd.DataFrame({'True y values':y,'Predicted y  values':y_pred})
```

```python
#preicting a single value
model.predict([[3790]])
```

```python
#predicting multiple values
model.predict([[2600],[2700],[2800]])
```

```python
#to check performance---if score is 1 then the model is best model
model.score(x,y)
```
